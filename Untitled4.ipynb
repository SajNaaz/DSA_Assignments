**3. Plot a bar graph of male vs female buyers participated in the sales**

Now let's check for the number of males and females participated in the sales event.

```python
male_numbr=dt[['Buyer Gender']].value_counts()['Male']
female_numbr=dt[['Buyer Gender']].value_counts()['Female']
print('The number of male and female s participated in event are',male_numbr ,female_numbr)
```

So 4948 males and 5052 females were participated.

```python
plt.figure(figsize=(15,8))
plot=plt.subplot(1,1,1)
male_numbr=dt[['Buyer Gender']].value_counts()['Male']
female_numbr=dt[['Buyer Gender']].value_counts()['Female']
plt.bar(dt['Buyer Gender']=='Male', male_numbr, width=0.5, align='center', label='Male')
plt.bar(dt['Buyer Gender']=='Female', female_numbr,width=0.4, align='edge', label='Female')
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(labels)
plt.xticks([])
plt.grid(linestyle='--')

plt.xlabel('Buyer Gender')
plt.ylabel('Sales Price')
plt.title('Bar graph of male vs female buyers participated in the sales')
plt.legend(loc='upper left')
plt.show()
```

**4. Find the top 5 cars based on their sales price.**


Let's sort out the top 5 cars in descending order and then seperate it out in a compressed table.

```python
top_cars=dt.sort_values('Sale Price', ascending=False).head()
top_cars
```

Compressed table consists of model and sales price value for the first 5 cars

```python
top_cars = top_cars[['Make','Model','Sale Price']]
top_cars.head()
```

So the top 5 cars having the best sale price are Volkswagen, Audi,Lexus,Mazda and Lexus ES model.

**5. Find the least 5 cars based on their Resell price.**
